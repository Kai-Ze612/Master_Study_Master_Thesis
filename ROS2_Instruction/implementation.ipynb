## How to Run
Set up the environment
```bash
cd /media/kai/Kai_Backup/Master_Study/Master_Thesis/Master_Study_Master_Thesis
source setup_environment.sh
```

Direct to the folder
```bash
cd /media/kai/Kai_Backup/Master_Study/Practical_Project/Practical_Courses/Master_Thesis/Master_Study_Master_Thesis/fr3_mujoco_ws
```

Create packages
```bash
./create_franka_phroject.sh
```

Add in setup.py
We should tell packages where is main function location, when we call python_node, it is equal to call main function.

We have to add main function under console_scripts
```bash
entry_points={
        'console_scripts': [
            'local = rl_remote_controller.local:main',
            'rl_remote = rl_remote_controller.rl_remote:main',
            'rl_training = rl_remote_controller.rl_training:main',
        ],
    },
```

Build the project  
```bash
colcon build --packages-select package_name
```

Run package setup script
```bash
source install/setup.bash
```

ROS2 Launch
```bash
ros2 launch franka_mujoco_controller position_control.launch.py
```
start idel status (include in launch.py file)
```bash
ros2 topic pub /start_push std_msgs/msg/String "{data: 'start'}"
```
